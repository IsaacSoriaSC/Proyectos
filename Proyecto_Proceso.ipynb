```
import numpy as np
from PIL import Image
import os
import matplotlib.pyplot as plt
import requests as rq
from bs4 import BeautifulSoup as bs
from wordcloud import WordCloud, ImageColorGenerator

#Creación de listas para su uso posterior.
etq = list() #etiquetas
vts = list() #votos que posee cada etiqueta
#Presentación del proyecto
print("UNIVERSIDAD DE GUAYAQUIL")
print("\n")
print("Proyecto del primer parcial: Nube de Palabras")
print("\n")
print('''Integrantes:
----->Bailón Guaranda Marcos
----->Calderón López Paul
----->Soria Cabrera Isaac\n''')
print("\n")
print("Curso:\nSOF - S- MA - 3 - 2")
print("\n")
print("Docente:\nIng. Miguel Botto Tobar")
print("\n")
print("El presente software mostrará una nube de palabras basada en los")  
print("términos más usadas por un usuario de Stackoverflow en español.")
print("\n")
print("Ingrese al siguiente link para la elección del usuario: https://es.stackoverflow.com/users/")
print("\n")
User= input("\n Ingrese el usario:  Ej:  7123/abulafia  ")
print("\n")
figure=0
while True:

  link='https://es.stackoverflow.com/users/'+User+'?tab=tags&sort=votes&page='+str(figure+1)

  if figure==0:
    print("\nEl link de la página es:"+" "+link+"\n")

  page=rq.get(link)
  soup=bs(page.content, 'html.parser')
  paginas=soup.find('div', class_='s-pagination--item is-selected')

  if paginas==None and figure==0:
    paginas=' '
  if paginas!=None: 
  
    try:

      tagTable=soup.find('table', class_='user-tags')
      etiq=tagTable.find_all('a', class_='post-tag') 
      votes=tagTable.find_all('div', class_='answer-votes') 
      # etiq=numero
      for x in etiq:
        etq.append(x.text)
      for x in votes:
        r=''
        if x.text=='0':
          vts.append(0.001)
        else:
          for y in x.text: 
            if y!='k' and y!='m': 
              r+=y 
            else:
              if y=='k':
                r=int(r)*10**3
              if y=='m':
                r=int(r)*10**6
          vts.append(int(r))
    except:
      print("User not found.")
    figure+=1

  else:
    break  

if etq!=[]:
  vtsZ=tuple(vts)

  #Creación de una variable que contenga los datos a proyectar en la nube de palabras.
  content=dict(zip(etq, vtsZ))
  #Creación de una nube de palabras personalizada usando máscaras.
  alice_mask = np.array(Image.open("alice_mask.png"))
  wc = WordCloud(background_color="white", width=400, height=300, mask=alice_mask, contour_width=3, contour_color='steelblue').generate_from_frequencies(content)
  fig = plt.figure()
  fig.set_figwidth(14) # set width
  fig.set_figheight(18) # set height
  plt.imshow(wc, interpolation='bilinear')
  plt.axis("off")
  plt.show()
```
